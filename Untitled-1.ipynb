# 2 Can you explain this behavior in which the antonyms are more similar than the synonyms?


Contextual Overlap:

 In many contexts, antonyms appear together more frequently than synonyms. This is because antonyms often define each other by contrast. For example, discussions about weight often include both "light" and "heavy" for comparison, making them contextually related in a way that "light" and "bright" might not be.

Polysemy of "Light":

 The word "light" has multiple meanings. It can refer to weight (the opposite of heavy) or illumination (related to bright). This polysemy might cause "light" to be more closely associated with "heavy" in contexts where weight is discussed.

Dimensional Salience:

 In vector space models like those used in word similarity computations, dimensions often represent salient features of words. "Light" and "heavy" might share more of these salient features (related to weight, density, etc.) than "light" and "bright" do, which could be more focused on luminosity and color.

Model Training and Data:

 The training data and the model's structure can also influence these results. If the model was trained on a dataset where "light" and "heavy" appear in similar contexts more frequently than "light" and "bright," this would affect their similarity scores.

```python
import gensim.downloader as api
import numpy as np

# Function to get top 10 neighbors of a word
def get_top_neighbors(model, word, topn=10):
    try:
        neighbors = model.most_similar(word, topn=topn)
        return [neighbor[0] for neighbor in neighbors]
    except KeyError:
        return []

# Function to calculate average similarity between neighbors from two models
def average_neighbor_similarity(model1, model2, neighbors1, neighbors2):
    similarities = []
    for n1 in neighbors1:
        for n2 in neighbors2:
            try:
                sim = model1.similarity(n1, n2)
                similarities.append(sim)
            except KeyError:
                continue
    return np.mean(similarities) if similarities else 0

# Function to process words and calculate similarities
def process_words(words, model1, model2):
    word_similarities = {}
    for word in words:
        neighbors_model1 = get_top_neighbors(model1, word)
        neighbors_model2 = get_top_neighbors(model2, word)
        avg_sim = average_neighbor_similarity(model1, model2, neighbors_model1, neighbors_model2)
        word_similarities[word] = avg_sim
    return word_similarities



# Define a list of words to test
words_to_test = [
    'love', 'money', 'music', 'happy', 'car', 'food', 'phone', 'internet', 
    'election', 'health', 'travel', 'sport', 'education', 'science', 'art', 
    'freedom', 'peace', 'culture', 'data', 'energy', 'environment', 'family', 
    'government', 'healthcare', 'history', 'job', 'law', 'media', 'nature', 
    'power', 'society', 'technology', 'university', 'war', 'water', 'work',
    'youth', 'beauty', 'communication', 'economy', 'ethics', 'fashion', 
    'global', 'language', 'leadership', 'marketing', 'network', 'quality', 
    'research', 'security', 'trend', 'value', 'wealth', 'wellness', 'wisdom'
]
# Process the words and get their similarities
similarities = process_words(words_to_test, glove_wiki, glove_twitter)

# Sort the words by their average similarity and print the top 5
sorted_words = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
top_5_words = sorted_words[:5]

for word, sim in top_5_words:
    print(f"Word: {word}, Average Neighbor Similarity: {sim}")
```

```python
import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


# Filter the first 5000 words, keeping only those ending with 'ed' or 'ing'
filtered_words = [word for word in model.index_to_key[:5000] if word.endswith('ed') or word.endswith('ing')]
word_vectors = [model[word] for word in filtered_words if word in model]

# Print the number of words filtered
print(f"Number of filtered words: {len(word_vectors)}")
# Applying PCA to reduce dimensions to 2
pca = PCA(n_components=2)
reduced_vectors = pca.fit_transform(word_vectors)
# Plotting the words
# Plotting the words
for word, coord in zip(filtered_words, reduced_vectors):
    color = 'blue' if word.endswith('ed') else 'green'
    plt.scatter(coord[0], coord[1], color=color)

plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Word Vectors Reduced to 2D with PCA')
plt.show()
```

# discussion of the plot

The scatter plot displays PCA-reduced word vectors for words ending in "ed" (blue) and "ing" (green). The key observations are:

There is substantial overlap between the two groups, indicating no clear separation.
This suggests that words with these suffixes do not differ enough in their usage or meaning to be distinctly apart in the reduced vector space.
The PCA reduction process, which is lossy, could have eliminated subtle differences that might be present in higher dimensions.
The similar distribution reflects the linguistic reality that words sharing the same root often appear in similar contexts, regardless of tense or aspect.
In essence, the lack of separation in the PCA plot highlights the complexity of natural language and the challenges of capturing the nuances of word forms with dimensionality reduction.

```python
def calculate_average_precision(relevancies):
    relevant_count = 0
    sum_precision = 0.0
    
    for i, is_relevant in enumerate(relevancies):
        if is_relevant:
            relevant_count += 1
            precision_at_i = relevant_count / (i + 1)
            sum_precision += precision_at_i
    
    average_precision = sum_precision / relevant_count if relevant_count > 0 else 0
    return average_precision
```
